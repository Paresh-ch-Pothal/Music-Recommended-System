```python
# Converting the list of the tags into string
new_df["tags"]=new_df["tags"].apply(lambda x: " ".join(x))
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(st):
    li=[]
    for i in st.split():
        li.append(ps.stem(i))
    return " ".join(li)
```

```python
new_df.head()
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
new_df["tags"]=new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df.iloc[0,4]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cs=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cs.fit_transform(new_df["tags"]).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cs.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])
```

```python
def recommendSongs(song_name):
    song_index=new_df[new_df["song"] == song_name].index[0]
    distances=similarity[song_index]
    songList=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in songList:
        print(new_df.iloc[i[0]].song)
```

```python
recommendSongs("As Good As New")
```
